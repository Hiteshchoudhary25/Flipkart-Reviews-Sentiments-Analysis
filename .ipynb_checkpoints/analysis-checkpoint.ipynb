# Flipkart Reviews Sentiment Analysis using Python

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

# from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
```

```python
data = pd.read_csv("https://raw.githubusercontent.com/amankharwal/Website-data/master/flipkart_reviews.csv")
print(data.head(5))
```

```python
data.isnull().sum()
```

# Sentiment Analysis

```python
import nltk
import re
nltk.download('stopwords')
stemmer = nltk.SnowballStemmer("english")
from nltk.corpus import stopwords
import string
stopword = set(stopwords.words('english'))

def clean(text):
    text = str(text).lower()
    text = re.sub(r'\[.*?\]', '', text)  # removes [text] blocks
    text = re.sub(r'https?://\S+|www\.\S+', '', text)  # removes URLs
    text = re.sub(r'<.*?>+', '', text)  # removes HTML tags
    text = re.sub(r'[%s]' % re.escape(string.punctuation), '', text)  # removes punctuation
    text = re.sub(r'\n', '', text)  # removes newlines
    text = re.sub(r'\w*\d\w*', '', text)  # removes words with digits
    text = [word for word in text.split(' ') if word not in stopword]  # removes stopwords
    text = " ".join(text)
    text = [stemmer.stem(word) for word in text.split(' ')]  # stemming
    text = " ".join(text)
    return text

data["Review"] = data["Review"].apply(clean)
```

```python
ratings = data["Rating"].value_counts()
ratings
```

```python
numbers = ratings.index
quantity = ratings.values
```

```python
import plotly.express as px
figure = px.pie(data, 
             values=quantity, 
             names=numbers,hole = 0.5,title="Distribution of Ratings")
figure.show()
```

```python
text = " ".join(i for i in data.Review)
stopwords = set(STOPWORDS)
wordcloud = WordCloud(stopwords=stopwords,background_color="white").generate(text)
plt.figure(figsize=(15,10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
nltk.download('vader_lexicon')
sentiments = SentimentIntensityAnalyzer()
data["Positive"] = [sentiments.polarity_scores(i)["pos"] for i in data["Review"]]
data["Negative"] = [sentiments.polarity_scores(i)["neg"] for i in data["Review"]]
data["Neutral"] = [sentiments.polarity_scores(i)["neu"] for i in data["Review"]]
data = data[["Review", "Positive", "Negative", "Neutral"]]
print(data.head())
```

```python
x = sum(data["Positive"])
y = sum(data["Negative"])
z = sum(data["Neutral"])

def sentiment_score(a, b, c):
    if (a>b) and (a>c):
        print("Positive üòä ")
    elif (b>a) and (b>c):
        print("Negative üò† ")
    else:
        print("Neutral üôÇ ")
sentiment_score(x, y, z)
```
